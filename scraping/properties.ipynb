```python
import requests
from bs4 import BeautifulSoup
```

```python
r1 = requests.get("http://www.pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
c1 = r1.content
sp1 = BeautifulSoup(c1, "html.parser")
```

```python
props = sp1.find_all("div", {"class": "propertyRow"})
cnt = 0
for prop in props:
    print(f"\n******** Property {cnt} ***********\n")
    price = prop.find('h4', {'class': 'propPrice'}).text.strip()
    print(price)
    address = prop.find_all('span', {'class': 'propAddressCollapse'})
    for add in address:
        print(add.text)
        
    infos = prop.find_all('div', {'class': 'secondaryDetails'})
    details, temp = [], []
    if infos:
        for info in infos:
            for detail in ([s for s in info.text.split('\n') if s]):
                print(detail)
    
    cnt += 1
#     print(prop)
```
